Let's define an array of angles and some trig functions.

```python
x = np.linspace(0, 2*np.pi, 100)
print(x[0],x[-1])

y = np.sin(x)
z = np.cos(x)
w = np.sin(4*x)
v = np.cos(4*x)
```

Let's make a two panel figure side by side.

```python
f, axarr = plt.subplots(1,2)
# 1 row by 2 columns figure

# first panel

axarr[0].plot(x,y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')

# second panel

axarr[1].plot(x,z)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

# add more space between the figures

f.subplots_adjust(wspace=0.4)
```

### OK, but the axis ratios are all squished. Let's change that.

```python
f, axarr = plt.subplots(1,2)
# 1 row by 2 columns figure

# first panel

axarr[0].plot(x,y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')

# second panel

axarr[1].plot(x,z)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

# add more space between the figures

f.subplots_adjust(wspace=0.4)

# fix the axis ratio

axarr[0].set_aspect('equal')
axarr[1].set_aspect(np.pi) #y:x
```

### Add a legend.

```python
# adjust size of figure

fig = plt.figure(figsize=(6,6))

plt.plot(x, y, label=r'$y = \sin(x)$') # add a label to y
plt.plot(x, z, label=r'$z = \cos(x)$') # add a label to y
plt.plot(x, w, label=r'$w = \sin(4x)$') # add a label to y
plt.plot(x, v, color="purple", label=r'$v = \cos(4x)$') # add a label to y

plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')

plt.xlim([0,2*np.pi])
plt.ylim([-1.2,1.2])
plt.legend(loc=1, framealpha=0.5)

# fix the axis ratio

plt.gca().set_aspect(np.pi/1.2)
```
